# 1. Text cleaning and preprocessing

```python
import spacy
nlp = spacy.load('en_core_web_sm')
import string
punctuation = string.punctuation
```

```python
def clean(text):
  #lowercasing

  cleaned_text = text.lower()

  #removing <>br<> character:

  cleaned_text = cleaned_text.replace("</br></br>"," ")
  cleaned_text = cleaned_text.replace("--"," ")


  #removing punctuation
  cleaned_text = cleaned_text.translate(str.maketrans('','',punctuation)) # syntax maketrans("argument.", "replace argument.", "remove a certain argument.")

  #removing digits:
  for char in cleaned_text:
    if char.isdigit():
      cleaned_text = cleaned_text.replace(char,"")

  #removing stopwords:



  return cleaned_text
```

```python
req_df.text = req_df.text.apply(lambda data: clean(data))
```
